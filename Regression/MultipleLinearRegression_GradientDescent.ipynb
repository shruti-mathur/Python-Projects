```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
data = pd.read_csv('data/Student.csv')
```

```python
math = data['Math'].values
read = data['Reading'].values
write = data['Writing'].values
```

```python
fig = plt.figure(figsize=(20,10))
ax = Axes3D(fig)
ax.scatter(math,read,write,color='lightcoral')
```

```python
n = len(math)
x0 = np.ones(n)
x = np.array([x0,math,read]).T
b = np.array([0,0,0])
y = np.array(write)
alpha = 0.0001
```

```python
def cost_function(x,y,b):
    m = len(y)
    j = np.sum((x.dot(b) - y)**2)/(2*m)
    
    return j
```

```python
cost_function(x,y,b)
```

```python
def gradient_descent(x,y,b,alpha,interval):
    cost_history = [0] * interval
    m = len(y)
    
    for i in range(interval):
        h = x.dot(b)
        loss = h - y
        gradient = x.T.dot(loss)/m
        
        b = b - alpha * gradient
        cost = cost_function(x,y,b)
        cost_history[i] = cost
        
        return b, cost_history
```

```python
newb, cost = gradient_descent(x,y,b,alpha,10000)
```
