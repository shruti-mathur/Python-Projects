```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
np.random.seed(0)
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
#df.head()
```

```python
print(iris)
```

```python
# Adding a new column for the species name ...
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Viewing the top 5 rows
df.head()
```

```python
# Creating test and train data
df['train'] = np.random.uniform(0,1, len(df)) <= 0.75    # generating random nos. between 0 and 1

df.head()
```

```python
# Creating dataframes with test rows  and training rows
train, test = df[df['train'] == True], df[df['train'] == False]

print('No.of observations in training data:', len(train))
print('No.of observations in testing data:', len(test))
```

```python
features = df.columns[:4]
```

```python
features
```

```python
y = pd.factorize(train['species'])[0]
```

```python
y
```

```python
# Creating a random forest classifier
clf = RandomForestClassifier(n_jobs= 2, random_state= 0)  # n_jobs --> changes your priority

# Training the classifier
clf.fit(train[features], y) # creating random forest classifier here ... y--> 0,1,2   
```

```python
# Applying trained classifier to the test
clf.predict(test[features])
```

```python
test[features]
```

```python
# Viewing the predicted probabilities of the first 10 observations ...
clf.predict_proba(test[features])[10:20]    # proba is for probability
```

```python
# mapping names for the plants for each  predicted  plant class
preds = iris.target_names[clf.predict(test[features])]

# View the PREDICTED  species for the first five observationds
preds[0:25]
```

```python
# Viewing actual  species for the first five observations

test['species'].head()
```

```python
# Creating confusion matrix
pd.crosstab(test['species'], preds, rownames=['Actual Species'], colnames=['Predicted Species'])
```
